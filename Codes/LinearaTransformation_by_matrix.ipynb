```python
import numpy as np
import matplotlib.pyplot as plt

# 선형변환을 수행할 행렬 정의
transformation_matrix = np.array([[2, 0], [0, 1]])

# 초기 벡터 집합 생성 (격자점)
x_values = np.linspace(-1, 1, 10)
y_values = np.linspace(-1, 1, 10)
grid = np.array(np.meshgrid(x_values, y_values))
points = grid.reshape(2, -1).T

# 행렬을 사용하여 벡터 변환
transformed_points = points @ transformation_matrix.T

# 원래의 벡터(격자점)와 변환된 벡터 시각화
plt.figure(figsize=(10, 5))

# 원래 벡터 플롯
plt.subplot(1, 2, 1)
plt.scatter(points[:, 0], points[:, 1], color='blue')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("원래 벡터")

# 변환된 벡터 플롯
plt.subplot(1, 2, 2)
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], color='red')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("변환된 벡터")

plt.show()
```

```python
# 회전, 스케일링, 전단 변환을 위한 행렬 정의
theta = np.radians(45)  # 45도 회전
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
scaling_matrix = np.array([[2, 0], [0, 2]])  # x와 y 방향으로 2배 확장
shear_matrix = np.array([[1, 1], [0, 1]])  # x축 방향으로 1의 전단

# 변환된 벡터 계산
rotated_points = points @ rotation_matrix.T
scaled_points = points @ scaling_matrix.T
sheared_points = points @ shear_matrix.T

# 원래 벡터 및 변환된 벡터 시각화
plt.figure(figsize=(15, 5))

# 회전 변환
plt.subplot(1, 3, 1)
plt.scatter(points[:, 0], points[:, 1], color='blue')
plt.scatter(rotated_points[:, 0], rotated_points[:, 1], color='red')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("회전 변환")

# 스케일링 변환
plt.subplot(1, 3, 2)
plt.scatter(points[:, 0], points[:, 1], color='blue')
plt.scatter(scaled_points[:, 0], scaled_points[:, 1], color='red')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("스케일링 변환")

# 전단 변환
plt.subplot(1, 3, 3)
plt.scatter(points[:, 0], points[:, 1], color='blue')
plt.scatter(sheared_points[:, 0], sheared_points[:, 1], color='red')
plt.xlim(-2, 4)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("전단 변환")

plt.show()
```
